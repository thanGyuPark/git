```python
# report_analysis.py
# -*- coding: utf-8 -*-

import os
import time
import json
import random
from typing import Dict, Optional, List, Tuple

import requests
import numpy as np
import pandas as pd
import yfinance as yf
from matplotlib import pyplot as plt
from pandas.tseries.offsets import DateOffset

from openai import OpenAI
from sec_api import ExtractorApi

# ---- Optional deps for RAG (Faiss 기반으로 변경) ----
import faiss
from sentence_transformers import SentenceTransformer


# =========================
# Config
# =========================
USE_CACHE = True
DEFAULT_LLM_MODEL = os.getenv("OPENAI_MODEL", "gpt-4o-mini")  # 필요하면 .env에서 변경
DEFAULT_EMBED_MODEL = os.getenv("EMBED_MODEL", "all-MiniLM-L6-v2")

# Projects 폴더 루트(원하면 바꿔도 됨)
PROJECTS_ROOT = os.getenv("PROJECTS_ROOT", "projects")


# =========================
# Helpers
# =========================
def post_with_backoff(url: str, payload: dict, max_retries: int = 6, timeout: int = 30) -> requests.Response:
    """
    429(레이트리밋) 대응: Retry-After 우선, 없으면 지수 백오프
    """
    last_err = None
    for i in range(max_retries):
        try:
            r = requests.post(url, json=payload, timeout=timeout)
            if r.status_code != 429:
                r.raise_for_status()
                return r

            retry_after = r.headers.get("Retry-After")
            wait = float(retry_after) if retry_after else min(2 ** i, 60) + random.random()
            time.sleep(wait)

        except Exception as e:
            last_err = e
            time.sleep(min(2 ** i, 30) + random.random())

    raise RuntimeError(f"SEC API request failed after retries. last_err={last_err}")


def safe_mkdir(path: str) -> None:
    os.makedirs(path, exist_ok=True)


# --- RAG 성능 향상을 위한 개선된 청킹 함수 (split_text_simple 대신 사용) ---
def split_text_recursive(text: str, chunk_size: int = 2000, overlap: int = 250) -> List[str]:
    """
    재귀적 분할 기반 청킹: 문단, 문장 등 의미 단위로 청킹하여 문맥을 보존
    """
    text = text or ""
    text = text.replace("\r\n", "\n")
    
    # 금융 보고서에 적합한 분할자: 문단, 문장, 쉼표 순
    separators = ["\n\n", "\n", ". ", " ", ""]
    
    def _split(t: str, sep_idx: int) -> List[str]:
        # 베이스 케이스: 청크 사이즈를 만족하거나, 분할자가 더 이상 없을 경우 단순 청킹
        if len(t) <= chunk_size or sep_idx >= len(separators) - 1:
            chunks = []
            start = 0
            n = len(t)
            while start < n:
                end = min(start + chunk_size, n)
                chunk = t[start:end].strip()
                if chunk:
                    chunks.append(chunk)
                start = end - overlap
                if start < 0:
                    start = 0
                if start >= n:
                    break
            return chunks

        separator = separators[sep_idx]
        if not separator:
            return _split(t, sep_idx + 1)

        parts = t.split(separator)
        
        final_chunks = []
        current_chunk = ""
        for part in parts:
            part_len = len(part)
            sep_len = len(separator) if current_chunk else 0
            
            if len(current_chunk) + sep_len + part_len <= chunk_size:
                # 청크 사이즈 내에 있으면 계속 합침
                if current_chunk:
                    current_chunk += separator
                current_chunk += part
            else:
                # 사이즈 초과 시, 현재 청크를 저장하고 새 청크 시작
                if current_chunk:
                    # 현재 청크가 너무 크면, 다음 분할자로 재귀 호출
                    sub_chunks = _split(current_chunk, sep_idx + 1)
                    final_chunks.extend(sub_chunks)
                
                current_chunk = part
                # 새 파트가 청크 사이즈를 초과하면 재귀 호출
                if len(current_chunk) > chunk_size:
                    sub_chunks = _split(current_chunk, sep_idx + 1)
                    final_chunks.extend(sub_chunks)
                    current_chunk = "" # 재귀 호출로 처리했으니 비움

        # 마지막으로 남아있는 current_chunk 처리
        if current_chunk:
            sub_chunks = _split(current_chunk, sep_idx + 1)
            final_chunks.extend(sub_chunks)
            
        return [c for c in final_chunks if c] # 빈 문자열 제거

    # 오버랩 처리를 위해 전체 텍스트를 한 번에 분할
    return _split(text, 0)
# -----------------------------------------------------------------


def plot_save(fig, path: str) -> str:
    safe_mkdir(os.path.dirname(path))
    fig.savefig(path, bbox_inches="tight")
    plt.close(fig)
    return path


# =========================
# Main Class
# =========================
class ReportAnalysis:
    def __init__(self, ticker_symbol: str):
        self.ticker_symbol = ticker_symbol.upper()

        # --- env keys ---
        self.openai_api_key = os.getenv("OPENAI_API_KEY")
        self.sec_api_key = os.getenv("SEC_API_KEY")
        if not self.openai_api_key:
            raise ValueError("OPENAI_API_KEY is missing. Put it in .env or environment variables.")
        if not self.sec_api_key:
            raise ValueError("SEC_API_KEY is missing. Put it in .env or environment variables.")

        # --- clients ---
        self.client = OpenAI(api_key=self.openai_api_key)
        self.extractor = ExtractorApi(self.sec_api_key)

        # --- market data ---
        self.stock = yf.Ticker(self.ticker_symbol)
        self.info = self.stock.info or {}

        # --- dirs ---
        self.project_dir = os.path.join(PROJECTS_ROOT, self.ticker_symbol)
        self.cache_dir = os.path.join(self.project_dir, "cache")
        safe_mkdir(self.project_dir)
        safe_mkdir(self.cache_dir)

        # --- report address (lazy) ---
        self.report_address: Optional[str] = None

        # --- embedder + Faiss (chroma 대체) ---
        self.embedder = SentenceTransformer(DEFAULT_EMBED_MODEL)
        # {섹션: (Faiss Index 객체, 문서 청크 리스트)}를 저장할 딕셔너리
        self.faiss_index: Dict[str, Tuple[faiss.IndexFlatL2, List[str]]] = {}
        
        # --- System Prompt ---
        self.system_prompt = """
Role: Expert Investor
Department: Finance
Primary Responsibility: Generation of Customized Financial Analysis Reports

Role Description:
As an Expert Investor within the finance domain, your expertise is harnessed to develop bespoke Financial Analysis Reports that cater to specific client requirements. This role demands a deep dive into financial statements and market data to unearth insights regarding a company's financial performance and stability.

Key Objectives:
- Analytical Precision
- Effective Communication
- Client Focus
- Adherence to Excellence
        """.strip()

    # -------------------------
    # SEC / 10-K
    # -------------------------
    def _ensure_report_address(self) -> str:
        """
        report_address를 필요할 때만 가져오도록(지연 로딩).
        캐시에 있으면 캐시 사용.
        """
        if self.report_address:
            return self.report_address

        address_json = os.path.join(self.project_dir, "sec_report_address.json")

        if USE_CACHE and os.path.exists(address_json):
            with open(address_json, "r", encoding="utf-8") as f:
                latest_10k_data = json.load(f)
            self.report_address = latest_10k_data.get("linkToFilingDetails")
            if self.report_address:
                return self.report_address

        # SEC-API 검색 호출 (429 백오프 포함)
        endpoint = f"https://api.sec-api.io?token={self.sec_api_key}"

        query = {
            "query": {
                "query_string": {
                    "query": f'ticker:{self.ticker_symbol} AND formType:"10-K"'
                }
            },
            "from": "0",
            "size": "1",
            "sort": [{"filedAt": {"order": "desc"}}],
        }

        response = post_with_backoff(endpoint, query)
        filings = response.json().get("filings", [])

        latest_10k_data = {"linkToFilingDetails": None}
        if filings:
            latest_10k_data = filings[0]

        with open(address_json, "w", encoding="utf-8") as f:
            json.dump(latest_10k_data, f, ensure_ascii=False, indent=2)

        self.report_address = latest_10k_data.get("linkToFilingDetails")
        if not self.report_address:
            raise RuntimeError(f"No 10-K filings found for {self.ticker_symbol} (or missing linkToFilingDetails).")
        return self.report_address

    def get_10k_section(self, section) -> str:
        """
        SEC EDGAR 10-K 섹션 텍스트 가져오기 (+ 캐시)
        """
        # 유효 섹션 목록 (list 대신 set 사용 권장)
        valid = [1, "1A", "1B", 2, 3, 4, 5, 6, 7, "7A", 8, 9, "9A", "9B", 10, 11, 12, 13, 14, 15]
        if section not in valid and str(section) not in valid:
            raise ValueError(f"Section must be in {valid}")

        section = str(section)
        safe_mkdir(os.path.join(self.project_dir, "10k"))
        report_name = os.path.join(self.project_dir, "10k", f"section_{section}.txt")

        if USE_CACHE and os.path.exists(report_name):
            with open(report_name, "r", encoding="utf-8") as f:
                return f.read()

        report_address = self._ensure_report_address()
        section_text = self.extractor.get_section(report_address, section, "text")

        with open(report_name, "w", encoding="utf-8") as f:
            f.write(section_text)

        return section_text

    # -------------------------
    # Financial statements
    # -------------------------
    def get_income_stmt(self) -> pd.DataFrame:
        return self.stock.financials

    def get_balance_sheet(self) -> pd.DataFrame:
        return self.stock.balance_sheet

    def get_cash_flow(self) -> pd.DataFrame:
        return self.stock.cashflow

    def get_company_info(self) -> Dict[str, str]:
        info = self.info or {}
        return {
            "Company Name": info.get("shortName", "N/A"),
            "Industry": info.get("industry", "N/A"),
            "Sector": info.get("sector", "N/A"),
            "Country": info.get("country", "N/A"),
        }

    def get_key_data(self) -> Dict[str, str]:
        hist = self.stock.history(period="6mo")
        info = self.info or {}

        if hist is None or hist.empty:
            return {"error": "No price history from yfinance."}

        close_price = float(hist["Close"].iloc[-1])
        avg_daily_volume_6m = float(hist["Volume"].mean())

        currency = info.get("currency", "USD")
        market_cap = info.get("marketCap", None)
        low = info.get("fiftyTwoWeekLow", None)
        high = info.get("fiftyTwoWeekHigh", None)
        book_value = info.get("bookValue", None)

        return {
            f"6개월 평균 일일 거래량 ({currency} 백만 단위)": f"{avg_daily_volume_6m / 1e6:.2f}",
            f"종가 ({currency})": f"{close_price:.2f}",
            f"시가총액 ({currency} 백만 단위)": f"{(market_cap / 1e6):.2f}" if market_cap else "N/A",
            f"52주 가격 범위 ({currency})": f"{low} - {high}",
            f"BVPS (주당 장부가치) ({currency})": f"{book_value}" if book_value is not None else "N/A",
        }

    # -------------------------
    # Charts
    # -------------------------
    def get_stock_performance(self) -> str:
        def fetch_close(ticker: str, period="1y") -> pd.Series:
            s = yf.Ticker(ticker)
            hist = s.history(period=period)
            return hist["Close"]

        target_close = fetch_close(self.ticker_symbol)
        sp500_close = fetch_close("^GSPC")

        company_change = (target_close - target_close.iloc[0]) / target_close.iloc[0] * 100
        sp500_change = (sp500_close - sp500_close.iloc[0]) / sp500_close.iloc[0] * 100

        start_date = company_change.index.min()
        four_months = start_date + DateOffset(months=4)
        eight_months = start_date + DateOffset(months=8)
        end_date = company_change.index.max()

        fig = plt.figure(figsize=(14, 7))
        plt.rcParams.update({"font.size": 14})

        name = self.info.get("shortName", self.ticker_symbol)
        plt.plot(company_change.index, company_change, label=f"{name} Change %")
        plt.plot(sp500_change.index, sp500_change, label="S&P 500 Change %")

        plt.title(f"{name} vs S&P 500 - Change % Over the Past Year")
        plt.xlabel("Date")
        plt.ylabel("Change %")
        plt.xticks(
            [start_date, four_months, eight_months, end_date],
            [start_date.strftime("%Y-%m"), four_months.strftime("%Y-%m"), eight_months.strftime("%Y-%m"), end_date.strftime("%Y-%m")],
        )
        plt.legend()
        plt.grid(True)

        plot_path = os.path.join(self.project_dir, "stock_performance.png")
        return plot_save(fig, plot_path)

    def get_pe_eps_performance(self) -> str:
        ss = self.get_income_stmt()
        if ss is None or ss.empty or "Diluted EPS" not in ss.index:
            raise RuntimeError("Income statement missing 'Diluted EPS' row.")

        eps_series = ss.loc["Diluted EPS", :].dropna()
        if eps_series.empty:
            raise RuntimeError("EPS series is empty.")

        historical_data = self.stock.history(period="5y")
        if historical_data is None or historical_data.empty:
            raise RuntimeError("No historical data from yfinance.")

        # eps index는 보통 회계연도 날짜(컬럼)라서 정렬 필요
        eps_dates = pd.to_datetime(eps_series.index).sort_values()

        close_map = {}
        for d in eps_dates:
            # asof: 해당 날짜 이전의 가장 가까운 시점 사용
            row = historical_data.asof(d)
            if row is None or (isinstance(row, pd.Series) and row.empty):
                continue
            close_map[d] = float(row["Close"])

        aligned_dates = []
        aligned_eps = []
        aligned_pe = []
        for d in eps_dates:
            if d in close_map:
                e = float(eps_series.loc[d])
                p = close_map[d]
                if e != 0:
                    aligned_dates.append(d)
                    aligned_eps.append(e)
                    aligned_pe.append(p / e)

        if not aligned_dates:
            raise RuntimeError("Could not align EPS dates with price history.")

        fig, ax1 = plt.subplots(figsize=(14, 7))
        plt.rcParams.update({"font.size": 14})

        ax1.set_xlabel("Date")
        ax1.set_ylabel("PE Ratio")
        ax1.plot(aligned_dates, aligned_pe)
        ax1.grid(True)

        ax2 = ax1.twinx()
        ax2.set_ylabel("EPS")
        ax2.plot(aligned_dates, aligned_eps)

        name = self.info.get("shortName", self.ticker_symbol)
        plt.title(f"{name} PE Ratios and EPS (aligned)")

        ax1.set_xticks(aligned_dates)
        ax1.set_xticklabels([d.strftime("%Y-%m") for d in aligned_dates], rotation=45)

        plt.tight_layout()
        plot_path = os.path.join(self.project_dir, "pe_eps_performance.png")
        return plot_save(fig, plot_path)

    # -------------------------
    # RAG (Faiss + SentenceTransformer)
    # -------------------------
    
    def _get_or_build_section_index(self, section: str) -> Tuple[faiss.IndexFlatL2, List[str]]:
        """
        Faiss 인덱스를 로드하거나 새로 생성하여 반환합니다.
        반환값: (Faiss Index 객체, 문서 청크 리스트)
        """
        section = str(section)
        
        # 1. 캐시 경로 설정
        index_path = os.path.join(self.cache_dir, f"faiss_index_{section}.bin")
        chunks_path = os.path.join(self.cache_dir, f"faiss_chunks_{section}.json")

        # 2. 인메모리 캐시 확인
        if section in self.faiss_index:
            return self.faiss_index[section]

        # 3. 디스크 캐시 로드
        if USE_CACHE and os.path.exists(index_path) and os.path.exists(chunks_path):
            try:
                # 인덱스 로드
                index = faiss.read_index(index_path)
                # 청크 로드
                with open(chunks_path, "r", encoding="utf-8") as f:
                    chunks = json.load(f)
                    
                self.faiss_index[section] = (index, chunks)
                return index, chunks
            except Exception as e:
                print(f"Faiss cache load failed for section {section}: {e}")
                # 로드 실패 시 재빌드를 위해 통과

        # 4. 인덱스 및 청크 빌드
        section_text = self.get_10k_section(section)
        
        # 개선된 재귀적 청킹 함수 사용
        chunks = split_text_recursive(section_text, chunk_size=2000, overlap=250)
        
        if not chunks:
            raise RuntimeError(f"Section {section} yielded no chunks.")

        # 임베딩 생성 (float32 numpy 배열로)
        embeddings = self.embedder.encode(chunks, normalize_embeddings=True)
        embeddings = embeddings.astype('float32')
        dimension = embeddings.shape[1]

        # Faiss 인덱스 생성 및 추가
        index = faiss.IndexFlatL2(dimension)
        index.add(embeddings)

        # 5. 디스크에 저장 및 인메모리 캐시 업데이트
        faiss.write_index(index, index_path)
        with open(chunks_path, "w", encoding="utf-8") as f:
            json.dump(chunks, f, ensure_ascii=False)
        
        self.faiss_index[section] = (index, chunks)
        return index, chunks


    def _embed(self, texts: List[str]) -> np.ndarray:
        # Faiss 호환을 위해 float32 numpy 배열로 반환
        vecs = self.embedder.encode(texts, normalize_embeddings=True)
        return vecs.astype('float32')

    
    def get_10k_rag_answer(self, section, question: str, k: int = 5) -> str:
        """
        섹션 문서 기반 RAG: (chunking -> embedding -> Faiss 저장) 후 질의
        """
        section = str(section)
        
        try:
            index, chunks = self._get_or_build_section_index(section)
        except RuntimeError as e:
             # 청크가 없을 경우 에러 메시지를 반환
            return f"Error building RAG index for section {section}: {e}"

        # 1. 질문 임베딩
        q_emb = self._embed([question])

        # 2. Faiss에서 유사성 검색 (k=5)
        # D: 거리, I: 인덱스
        D, I = index.search(q_emb, k)
        
        # 3. 검색된 문서 청크 추출
        retrieved_indices = I[0]
        context_docs = [chunks[i] for i in retrieved_indices if i >= 0]
        context = "\n\n---\n\n".join(context_docs)

        rag_prompt = f"""
{self.system_prompt}

You are given CONTEXT extracted from the company's 10-K section {section}.
Answer the QUESTION using ONLY the context. If the context is insufficient, say what is missing explicitly.

CONTEXT:
{context}

QUESTION:
{question}
""".strip()

        # 4. LLM 호출
        chat = self.client.chat.completions.create(
            model=DEFAULT_LLM_MODEL,
            messages=[{"role": "user", "content": rag_prompt}],
            temperature=0,
            max_tokens=700,
        )
        return chat.choices[0].message.content

    # -------------------------
    # LLM Q&A (RAG / non-RAG)
    # -------------------------
    def ask_question(self, question: str, section=7, table_str: Optional[str] = None, use_rag: bool = False) -> str:
        """
        use_rag=True면 10-K 섹션 기반 RAG로 답변.
        use_rag=False면 섹션 텍스트를 그대로 Resource로 넣어서 답변.
        """
        if use_rag:
            return self.get_10k_rag_answer(section=section, question=question, k=5)

        section_text = self.get_10k_section(section)

        if table_str:
            prompt = f"{self.system_prompt}\n\n{table_str}\n\nResource:\n{section_text}\n\nQuestion:\n{question}"
        else:
            prompt = f"{self.system_prompt}\n\nResource:\n{section_text}\n\nQuestion:\n{question}"

        chat = self.client.chat.completions.create(
            model=DEFAULT_LLM_MODEL,
            messages=[{"role": "user", "content": prompt.strip()}],
            temperature=0,
            max_tokens=700,
        )
        return chat.choices[0].message.content

    # -------------------------
    # Analyses
    # -------------------------
    def analyze_income_stmt(self) -> str:
        cache_answer = os.path.join(self.project_dir, "income_stmt_analysis.txt")
        if USE_CACHE and os.path.exists(cache_answer):
            return open(cache_answer, "r", encoding="utf-8").read()

        income_stmt = self.get_income_stmt()
        df_string = "Income statement:\n" + income_stmt.to_string().strip()

        question = (
            "Embark on a thorough analysis of the company's income statement for the current fiscal year, "
            "focusing on revenue streams, cost of goods sold, operating expenses, and net income to discern "
            "the operational performance and profitability. Examine gross profit margin, operating margin, "
            "and net profit margin. Compare against historical data to identify trends. "
            "Conclude with a strategic overview in a single paragraph. Less than 130 words."
        )

        answer = self.ask_question(question, section=7, table_str=df_string, use_rag=False)
        with open(cache_answer, "w", encoding="utf-8") as f:
            f.write(answer)
        return answer

    def analyze_balance_sheet(self) -> str:
        cache_answer = os.path.join(self.project_dir, "balance_sheet_analysis.txt")
        if USE_CACHE and os.path.exists(cache_answer):
            return open(cache_answer, "r", encoding="utf-8").read()

        balance_sheet = self.get_balance_sheet()
        df_string = "Balance sheet:\n" + balance_sheet.to_string().strip()

        question = (
            "Delve into a detailed scrutiny of the company's balance sheet for the most recent fiscal year, "
            "pinpointing the structure of assets, liabilities, and shareholders' equity to decode financial "
            "stability. Evaluate liquidity, solvency, and equity position. Contrast with prior years. "
            "Conclude with a strategic assessment in a single paragraph. Less than 130 words."
        )

        answer = self.ask_question(question, section=7, table_str=df_string, use_rag=False)
        with open(cache_answer, "w", encoding="utf-8") as f:
            f.write(answer)
        return answer

    def analyze_cash_flow(self) -> str:
        cache_answer = os.path.join(self.project_dir, "cash_flow_analysis.txt")
        if USE_CACHE and os.path.exists(cache_answer):
            return open(cache_answer, "r", encoding="utf-8").read()

        cash_flow = self.get_cash_flow()
        df_string = "Cash flow:\n" + cash_flow.to_string().strip()

        question = (
            "Dive into a comprehensive evaluation of the company's cash flow for the latest fiscal year, "
            "focusing on operating, investing, and financing activities. Compare to prior periods. "
            "Conclude with a single-paragraph assessment. Less than 130 words."
        )

        answer = self.ask_question(question, section=7, table_str=df_string, use_rag=False)
        with open(cache_answer, "w", encoding="utf-8") as f:
            f.write(answer)
        return answer

    def financial_summarization(self) -> Dict[str, str]:
        income_stmt_analysis = self.analyze_income_stmt()
        balance_sheet_analysis = self.analyze_balance_sheet()
        cash_flow_analysis = self.analyze_cash_flow()

        cache_answer = os.path.join(self.project_dir, "financial_summarization.txt")
        if USE_CACHE and os.path.exists(cache_answer):
            summary = open(cache_answer, "r", encoding="utf-8").read()
        else:
            question = (
                f"Income statement analysis: {income_stmt_analysis}\n\n"
                f"Balance sheet analysis: {balance_sheet_analysis}\n\n"
                f"Cash flow analysis: {cash_flow_analysis}\n\n"
                "Synthesize the findings. Highlight core insights, risks, and opportunities. "
                "Offer recommendations to enhance shareholder value in a single paragraph. Less than 150 words."
            )
            summary = self.ask_question(question, section=7, use_rag=False)
            with open(cache_answer, "w", encoding="utf-8") as f:
                f.write(summary)

        return {
            "Income Statement Analysis": income_stmt_analysis,
            "Balance Sheet Analysis": balance_sheet_analysis,
            "Cash Flow Analysis": cash_flow_analysis,
            "Financial Summary": summary,
        }


# =========================
# Example usage
# =========================
if __name__ == "__main__":
    # 예: python report_analysis.py
    # 종목은 환경 변수 TICKER를 사용하거나, 'AAPL'을 기본값으로 사용
    ticker_symbol = os.getenv("TICKER", "AAPL")
    print(f"--- Starting analysis for Ticker: {ticker_symbol} ---")

    try:
        ra = ReportAnalysis(ticker_symbol)

        print("\n## 1. Company Info & Key Data")
        print(ra.get_company_info())
        print(ra.get_key_data())

        print("\n## 2. Chart Generation")
        print("Stock Performance Chart saved:", ra.get_stock_performance())
        print("PE/EPS Performance Chart saved:", ra.get_pe_eps_performance())

        print("\n## 3. Financial Analysis Summarization (LLM)")
        financial_results = ra.financial_summarization()
        
        print("\n### Income Statement Analysis")
        print(financial_results["Income Statement Analysis"])
        
        print("\n### Balance Sheet Analysis")
        print(financial_results["Balance Sheet Analysis"])
        
        print("\n### Cash Flow Analysis")
        print(financial_results["Cash Flow Analysis"])
        
        print("\n### Overall Financial Summary")
        print(financial_results["Financial Summary"])

        # RAG 테스트 (10-K Section 1A: Risk Factors)
        print("\n## 4. RAG Query Test (10-K Section 1A)")
        rag_question = "What are the primary competitive risks the company faces according to its latest 10-K filing?"
        rag_answer = ra.get_10k_rag_answer(section="1A", question=rag_question, k=5)
        print(f"Question: {rag_question}")
        print(f"RAG Answer: {rag_answer}")

    except Exception as e:
        print(f"\n!!! Analysis failed: {e} !!!")
```
